#Criando um Dataframe e Renomeando

---



# Importar a Biblioteca  

```python
#importando a biblioteca
import pandas as pd
```

# Criando um DataFrame





```python
alunos_df = pd.DataFrame({
    'nome': ['Aya', 'John', 'Denis'],
    'sexo': ['M', 'F', 'M'],
    'idade': [22, 26, 28]
})
```

```python
alunos_df
```

```python
alunos_df.info()
```

#Renomenado as colunas de um DataFrame

```python
alunos_df.columns # Lista o nome de todas as colunas do DataFrame
```

```python
list(alunos_df.columns) #converter para uma Lista
```

```python
alunos_df_renomeado= alunos_df.rename(columns={
    'nome':'Nome do Aluno',
    'idade': 'Idade' # Renomeia a coluna 'idade' para 'Idade'
})
```

```python
alunos_df_renomeado
```

```python
alunos_df
```

Note que o parâmetro rename cria cópia e não altera a tabela original(alunos_df), para alterar de forma definitiva utilizamos o método inplace= True

```python
alunos_df.rename(columns={
    'nome':'Nome do Aluno',
    'idade': 'Idade' # Renomeia a coluna 'idade' para 'Idade de forma definitiva'
}, inplace=True)
```

```python
alunos_df
```

```python
alunos_df.columns= ['NOME','SEXO','IDADE'] #Renomear todas as colunas de uma vez
```
