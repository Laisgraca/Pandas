#Series
Array uni-dimensional com dados e rótulos de um eixo

```python
import pandas as pd
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/Laisgraca/Pandas/refs/heads/main/gasprices.csv', sep='\t')
```

```python
data.head()
```

```python
data['ESTADO'] # Selecionando uma coluna inteira
```

```python
# Outra forma de selecionar a coluna inteira
# Só funciona para colunas com nomes sem espaço,acentos e etc)

data.ESTADO
```

```python
type(data['ESTADO']) # Informa o tipo de dados
```

```python
data.iloc[4] # Selecionando as informações contidas no índice [4] do dataframe
```

```python
type(data.iloc[4])
```
