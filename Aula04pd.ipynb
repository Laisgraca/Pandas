#Atribuindo Dados

```python
import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/Laisgraca/Pandas/refs/heads/main/gasprices.csv', sep='\t')
data.head()
```

#Atribuindo constantes

```python
produto_view = data['PRODUTO'] # A series retornada à coluna, Não é uma cópia, mas apenas uma visualização referência/view
produto_view
```

#Gerar Cópia

```python
produto_copia_bkp = data['PRODUTO'].copy() # Retorna uma cópia da coluna 'PRODUTO'
produto_copia_bkp
```

#Atribuindo o valor constante

```python
data['PRODUTO']= 'Combustível' # Atribuindo o valor constante ' Combustível' para linha do dataframe na coluna ' PRODUTO'
data.head()
produto_view # Como é uma visualização ele também sofre alteração quando os dados são modificados
```

```python
produto_copia_bkp # Mantém o valor original diferente do view
```

#Atribuindo listas ou series

```python
nrows, ncols = data.shape
nrows, ncols
```

```python
novos_produtos = [f'Produto {i} ' for i in range(nrows)]
novos_produtos # Mudando cada linha da coluna produto para que fique númerica de acordo com o índice
```

```python
len(novos_produtos) # Tamanho da lista
```

```python
# A quantidade de elementos da lista 'novos_produtos' é igual ao número de linhas do dataframe
data['PRODUTO'] = novos_produtos
data
```

```python
# voltando para os dados originais
data['PRODUTO']= produto_copia_bkp # produto_copia_bkp é uma Series
data.head()
```
