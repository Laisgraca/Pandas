#Criando Novas Colunas

```python
import pandas as pd
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/Laisgraca/Pandas/refs/heads/main/gasprices.csv', sep='\t')
data.head()
```

```python
# Criando uma coluna a partir de um valor constante/default
# Todas as linhas terão o mesmo valor para esta nova coluna
data['Posto'] = 'DEFAULT'
data
```

```python
# Não funciona pq a quantidade de elementos da lista (a serem atribuídos a nova coluna) é diferente
# da quantidade de linhas do dataframe
data['não funciona'] = [1,2,3]
```

```python
data['coluna a partir de lista'] = range(data.shape[0])
data
```

```python
# Criar uma coluna a partir de uma existente
data['PREÇO MÉDIO REVENDA (dólares)'] = data['PREÇO MÉDIO REVENDA'] * 5.08
data
```
